```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import rc, rcParams
%matplotlib inline
```

```python
# figure 1: semi-supervised
x = np.array([20, 80, 160, 320])
umt = np.array([31.47, 31.47, 31.47, 31.47])
semi = np.array([30.01, 36.80, 38.92, 40.16])
semi_umt = np.array([36.95, 38.73, 38.66, 38.67])
mt = np.array([39.49, 39.49, 39.49, 39.49])

SMALL_SIZE = 14
MEDIUM_SIZE = 15
BIGGER_SIZE = 17

plt.rc('font', size=SMALL_SIZE)          # controls default text sizes
plt.rc('axes', titlesize=SMALL_SIZE)     # fontsize of the axes title
plt.rc('axes', labelsize=MEDIUM_SIZE)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize=SMALL_SIZE)    # legend fontsize
plt.rc('figure', titlesize=BIGGER_SIZE) 
plt.rcParams['axes.labelweight'] = 'bold'
rc('font', weight='bold')
rcParams['figure.figsize'] = 5, 6.5
linewidth = 3.
plt.plot(x, umt, marker='o', linewidth=linewidth)
plt.plot(x, semi, marker='^', linewidth=linewidth)
plt.plot(x, semi_umt, marker='s', linewidth=linewidth)
plt.plot(x, mt, marker='p', linewidth=linewidth)
plt.xticks(np.arange(0, 360, 40.0))
plt.yticks(np.arange(29, 43, 1.0))
plt.xlabel('In-Domain Size (K)')
plt.ylabel('Validation BLEU')
plt.legend(['IBT', 'SEMI', 'IBT+SEMI', 'MT (Sup.)'], bbox_to_anchor=(0.9, 0.65), loc=1, borderaxespad=0.)
plt.savefig('semi.png', dpi=1200)
plt.show()
```

```python
# figure 2: vary out-of-domain size
x = np.array([0.1, 0.5, 1, 4])
unadapt = np.array([7.64, 10.46, 12.02, 11.74])
bt = np.array([25.98, 26.89, 28.00, 29.58])
umt = np.array([41.34, 41.34, 41.34, 41.34])
umt_bt = np.array([48.55, 54.32, 55.65, 55.06])

SMALL_SIZE = 14
MEDIUM_SIZE = 15
BIGGER_SIZE = 17

plt.rc('font', size=SMALL_SIZE)          # controls default text sizes
plt.rc('axes', titlesize=SMALL_SIZE)     # fontsize of the axes title
plt.rc('axes', labelsize=MEDIUM_SIZE)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize=SMALL_SIZE)    # legend fontsize
plt.rc('figure', titlesize=BIGGER_SIZE) 
plt.rcParams['axes.labelweight'] = 'bold'
rc('font', weight='bold')
rcParams['figure.figsize'] = 5, 6.5
linewidth = 3.
plt.plot(x, unadapt, marker='o', linewidth=linewidth)
plt.plot(x, bt, marker='^', linewidth=linewidth)
plt.plot(x, umt, marker='s', linewidth=linewidth)
plt.plot(x, umt_bt, marker='p', linewidth=linewidth)
plt.xticks([0.1, 0.5, 1, 4])
plt.yticks(np.arange(4, 62, 4.0))
plt.xlabel('Out-of-Domain Size (M)')
plt.ylabel('Validation BLEU')
plt.legend(['Unadapted', 'Back', 'IBT', 'IBT+Back'], bbox_to_anchor=(0.9, 0.42), loc=1, borderaxespad=0.)
plt.savefig('vary_out_domain.png', dpi=1200)
plt.show()
```
